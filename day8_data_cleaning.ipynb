```python
import pandas as pd
```

```python
data = {
    "Name": ["John", "Mary", "Ali", "Ngozi", "Mary"],
    "Age": [25, None, 33, 29, 40],
    "Temperature": [36.5, 38.2, None, 39.1, 38.2],
    "Diagnosis": ["Malaria", "Typhoid", "Healthy", None, "Typhoid"]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
clean_up= df.dropna()
```

```python
print(clean_up)
```

```python
clean_df = df.drop_duplicates()
print(clean_df)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
print(df)
```

The dataset contained missing values.
Missing ages were replaced with the mean age.
Duplicate records were identified.
