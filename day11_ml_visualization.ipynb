```python
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = {
    "Age": [20, 25, 30, 35, 40, 45, 50],
    "Blood_Pressure": [110, 115, 120, 130, 135, 145, 150]
}

df = pd.DataFrame(data)
```

```python
x = df[["Age"]]
y = df["Blood_Pressure"]
```

```python
model = LinearRegression()

model.fit(x, y)
```

```python
predictions = model.predict(x)
print(predictions)
```

```python
plt.scatter(df["Age"], df["Blood_Pressure"])

plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Age Vs Blood Pressure")
plt.show()
```

```python
plt.scatter(df["Age"], df["Blood_Pressure"])

plt.plot(df["Age"], predictions)

plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Regression Line")

plt.show()
```

```python
new_prediction = model.predict([[60]])

print(new_prediction)
```

BP increased with age.
Regression line showed positive relationship.
Model predicted higher BP for older patients.
