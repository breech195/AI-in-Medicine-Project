```python
import pandas as pd
```

```python
data = {
    "Name": ["John", "Mary", "Ali", "Ngozi", "David", "Sarah"],
    "Age": [25, 40, 33, 29, 50, 21],
    "Diagnosis": ["Malaria", "Typhoid", "Malaria", "Hypertension", "Typhoid", "Malaria"],
    "Blood_Pressure": [120, 150, 130, 145, 160, 110]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
sorted_df = df.sort_values("Age")
```

```python
print(sorted_df)
```

```python
sorted_df = df.sort_values("Age", ascending=False)
```

```python
print(sorted_df)
```

```python
grouped = df.groupby("Diagnosis")
```

```python
print(grouped)
```

```python
print(df.groupby("Diagnosis").size())
```

```python
print(df.groupby("Diagnosis")["Blood_Pressure"].mean())
```

Malaria was most common diagnosis.
Typhoid patients had higher average BP.
Youngest patient was 21.
