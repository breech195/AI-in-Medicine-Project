```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["John", "Mary", "Ali", "Ngozi", "David"],
    "Age": [25, 30, 34, 19, 50],
    "Temperature": [35.5, 36.0, 38.2, 40.7, 37.6],
    "Diagnosis": ["Malaria", "Healthy", "Cancer", "Malaria", "Diabetes"],
}
df = pd.DataFrame(data)
```

```python
plt.bar(df["Name"], df["Age"])

plt.title("Age of Patients")
plt.xlabel("Patient name")
plt.ylabel("Age")

plt.show()
```

```python
plt.plot(df["Name"], df["Temperature"])

plt.title("Pateint Temperatures")
plt.xlabel("patient Name")
plt.ylabel("Patient Temperature")
```

```python
diagnosis_counts = df["Diagnosis"].value_counts()
```

```python
print(diagnosis_counts)
```

```python
diagnosis_counts.plot(kind="bar")

plt.title("Diagnosis Frequency")
plt.xlabel("Diagnosis")
plt.ylabel("Count")

plt.show()
```
