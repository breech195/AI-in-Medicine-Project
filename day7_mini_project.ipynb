```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["John", "Mary", "Ali", "Ngozi", "David", "Sarah"],
    "Age": [15, 65, 34, 21, 43, 21],
    "Temperature": [36.5, 38.2, 37.0, 39.1, 36.8, 38.5],
    "Blood Pressure": [120, 150, 130, 145, 160, 110],
    "Diagnosis": ["Malaria", "Typhoid", "Healthy", "Malaria", "Hypertension", "Typhoid"]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
print(df["Age"].mean())
```

```python
fever_patients = df[df["Temperature"] > 37.5]

print(fever_patients)
```

```python
diagnosis_counts = df["Diagnosis"].value_counts()
```

```python
diagnosis_counts.plot(kind="bar")

plt.title("Diagnosis Frequency")
plt.xlabel("Diagnosis")
plt.ylabel("Count")

plt.show()
```

Interpretation:
Malaria and typhoid are the most common diseases
3 persons were found to be hypertensive
Highest blood pressure is 160
3 persons have fever
The average age is 33.2
3 persons have fever above 37.5
