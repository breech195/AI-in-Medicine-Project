```python
import pandas as pd
```

```python
data = {
    "Name": ["John", "Mary", "Ali", "Ngozi", "David"],
    "Age": [25, 30, 42, 32, 18],
    "Temperature": [36.0, 38.5, 39.7, 40.0, 35.8],
    "Diagnosis": ["Malaria", "Hypertension", "Diabetes", "Healthy", "Cancer"]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
print(df["Age"].mean())
```

```python
print(df["Temperature"].min())
```

```python
print(df["Age"].min())
```

```python
print(df["Diagnosis"])
```

```python
fever_patients = df[df["Temperature"]>37.5]
```

```python
print(fever_patients)
```

```python
df["Risk"] = ["Low", "High", "Low", "High", "Medium"]
print(df)
```

```python
df.to_csv("patient_data.csv", index=False)
```
